```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap = None):
  fig = plt.figure(figsize = (12,10))
  ax = fig.add_subplot(111)
  ax.imshow(img, cmap = 'gray')
```

```python
reeses = cv2.imread('/reeses_puffs.png', 0)
display(reeses)
```

```python
cereals = cv2.imread('/many_cereals.jpg', 0)
```

```python
display(cereals)
```

```python
# Initiate ORB Detector
```

```python
orb = cv2.ORB_create()

kp1, des1 = orb.detectAndCompute(reeses, None)
kp2, des2 = orb.detectAndCompute(cereals, None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)

matches = bf.match(des1, des2)

matches = sorted(matches, key = lambda x: x.distance)

reeses_matches = cv2.drawMatches(reeses, kp1, cereals, kp2, matches[:25], None, flags = 2)
```

```python
display(reeses_matches)
```

```python
# Initiate SIFT detector
```

```python
sift = cv2.xfeatures2d.SIFT_create()

kp1, des1 = sift.detectAndCompute(reeses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)

PLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = PLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)

flann = cv2.FlannBasedMatcher(index_params, search_params)
matches = flann.knnMatch(des1, des2, k = 2)

matchesMask = [[0, 0] for i in range(len(matches))]
```

```python
for i, (match1, match2) in enumerate(matches):
  if match1.distance < 0.7 * match2.distance:
    matchesMask[i] = [1, 0]
```

```python
draw_params = dict(matchColor = (0, 255, 0), singlePointColor = (255, 0, 0), matchesMask = matchesMask, flags = 0)
```

```python
draw_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, matches, None, **draw_params)
```

```python
display(draw_matches)
```
