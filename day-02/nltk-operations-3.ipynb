```python
# Import spacy and language module from spacy.language
```

```python
import spacy
from spacy.language import Language
```

```python
# Load the language model
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
# Set the language component
```

```python
@Language.component("set_custom_boundaries")
def set_custom_boundaries(doc):
    for token in doc[:-1]:
        if token.text == ';':
            doc[token.i + 1].is_sent_start = True
    return doc
```

```python
# Add a pipeline
```

```python
nlp.add_pipe("set_custom_boundaries", before="parser")
```

```python
# To print all pipeline names
```

```python
print(nlp.pipe_names)
```

```python
# Define a NLP data object
```

```python
doc = nlp("'Management is doing things right; leadership is doing the right things.' - Peter Drucker")
```

```python
# Print the lines separated based on semicolons
# NOTE: We use language component to set boundary to treat semicolon as the end of sentence
```

```python
for sent in doc.sents:
    print(sent)      # [or] print(sent.text)
```
