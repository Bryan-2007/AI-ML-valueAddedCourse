```python
# Import spacy package and load nlp
```

```python
import spacy
from spacy.matcher import Matcher
nlp = spacy.load('en_core_web_sm')
```

```python
# Initiate matcher object for finding matches
```

```python
matcher = Matcher(nlp.vocab)
```

```python
# Initiate the patterns
```

```python
p1 = [{'LOWER' : 'solarpower'}]
p2 = [{'LOWER' : 'solar'}, {'LOWER' : 'power'}]
p3 = [{'LOWER' : 'solar'}, {'IS_PUNCT' : True}, {'LOWER' : 'power'}]
```

```python
# Add the patterns to matcher object with a name 'new_mat'
```

```python
matcher.add('new_mat', [p1, p2, p3])
```

```python
# Create a nlp data
```

```python
doc = nlp(u"The future of energy is the solar-power. Solar energy depends only on the sun and its called solarpower")
```

```python
# Initialize object to store the matches
```

```python
matches = matcher(doc)
print(matches)
```

```python
# Remove the defined pattern 'new_mat'
```

```python
matcher.remove('new_mat')
```

```python
# Display as diagram using render with colours- part 2
```

```python
doc4 = nlp(u"This is a possibly another longer sentance. This is important although it is not very important")

spans = list(doc4.sents)

# visualization options
options = {"distance": 110, "compact": False, "color": "yellow", "bg": "red"}

# render dependency diagram
displacy.render(spans, style="dep", options=options, jupyter=True)
```

```python
# Defining a function to print entities of a NLP object
```

```python
def show_ents(d):
    if d.ents:
        for i in d.ents:
            print(i.text + '-' + i.label_ + '-' + str(spacy.explain(i.label_)))
    else:
        print("No named entities found")
```

```python
# Printing the entities of a NLP object using the pre-defined function
```

```python
doc5 = nlp("Apple was founded by Steve Jobs in California in 1976")
show_ents(doc5)
```

```python
# Import span from spacy.tokens
```

```python
from spacy.tokens import Span
```

```python
# To clear existing entities
```

```python
doc5.ents = []
```

```python
# Create a new entity for the first token of doc5 - custom made entity
```

```python
org = doc5.vocab.strings['ORG']
new_ent = Span(doc5, 0, 1, label=org)

doc5.ents = [new_ent]
show_ents(doc5)
```
