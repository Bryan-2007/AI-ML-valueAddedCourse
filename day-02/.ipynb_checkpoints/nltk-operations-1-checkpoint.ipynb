```python
# Import Natural Language Tool Kit - NLTK
```

```python
import nltk

from nltk.stem.porter import *
```

```python
# Create object for porter stemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
# Defining an array of words
```

```python
words = ['hi', 'hello', 'bye', 'Hallo', 'hey', 'Tschuss']
```

```python
# It gives the stem word (fundamental word of the given word / base word)
for i in words:
    print(i, '--->', p_stemmer.stem(i))
```

```python
# Importing snowball stemmer
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
# Assign language as english for snowball stemmer
```

```python
s_stemmer = SnowballStemmer(language = 'english')
```

```python
# Finding the stem words of a sentence
```

```python
sentence = 'I am batman. I roam into the darkness fighting against the dark forces of the night'
for i in sentence.split():
    print(i , '------>', s_stemmer.stem(i))
```

```python
# Import spaCy package
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
text =  nlp(u"Natural Language Processing (NLP) is a field of Artificial Intelligence that helps computers understand human language")
```

```python
for i in text:
    print(f"{i.text:<17} {i.pos_:<10} {i.lemma:<27} {i.lemma_}")
```
