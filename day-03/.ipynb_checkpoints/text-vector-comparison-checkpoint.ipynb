```python
# Import spacy and download language model
```

```python
import spacy
spacy.cli.download("en_core_web_md")
```

```python
# Load the labguage model
```

```python
nlp = spacy.load("en_core_web_md")
```

```python
# Print the vector form of the NLP object - part 1
```

```python
nlp(u"lion").vector
```

```python
# Print the vector form of NLP object - part 2
```

```python
doc = nlp(u"The quick brown fox jumped over the lazy dogs")
doc.vector
```

```python
# Compare tokens of NLP object using similarity
```

```python
tokens = doc
for t1 in tokens:
    for t2 in tokens:
        print(t1.text,t2.text,t1.similarity(t2))
```

```python
# To find similarity between two NLP objects
```

```python
nlp(u"Lion").similarity(nlp(u"Dog"))
```
