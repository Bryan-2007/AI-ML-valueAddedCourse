```python
# Importing spaCy module
```

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
# Define nlp of a sentence
```

```python
doc = nlp(u'Elon Musk bought Twitter for $5 million and renamed it to "X"')
```

```python
# Access the tokens with info
# Text - The data
# POS - Parts of speech
# Dep - Dependency
# Lemma - Lemmatization
# Tag - Detailed POS
```

```python
print("Text\t\tPOS\t\tDependency\tLemmatization")
for token in doc:
    print(token.text + '\t\t' + token.pos_ + '\t\t' + token.dep_ + '\t\t' + token.lemma_)
```

```python
# To print the components of nlp (pipeline - dataflow)
```

```python
print(nlp.pipeline)
```

```python
# Print only names of the components
```

```python
print(nlp.pipe_names)
```

```python
# Get first word of the data
```

```python
print(doc[0])
```

```python
# Get the type of data
```

```python
print(type(doc))
```

```python
# Print the parts of speech of 5th word
```

```python
print(doc[4].pos_)
```

```python
# To print detailed POS of a word
```

```python
print(doc[0].tag_)
```

```python
# More Functions
# is_alpha --> checks whether word only contains letters
# is_stop --> checks whether word is a stop word ('is', 'and', 'the')
```

```python
print(doc[0].is_alpha)
print(doc[0].is_stop)
```

```python
# Accessing parts of the data as words
```

```python
doc1 = nlp(u"On 23-02-2026, CompanyX recorded ₹1,25,000 in revenue, a 15% increase from 2025. The CEO emailed users at 10:30 AM, saying \"Launch v2.0 is live!\" Visit www.companyx.com or call +91-98765-43210 for support. Errors like #404 and @timeout were resolved within 24-hrs.")

print(doc1[10], '\n')
print(doc1[7:14], '\n')
print(doc1[-14:])
```

```python
# Printing all words in the nlp dataset
```

```python
for i in doc1:
    print(i)
```

```python
# Print length of the nlp set and vocab set
```

```python
print(len(doc1))
print(len(doc1.vocab))
```

```python
# To create a named entity
```

```python
for i in doc1.ents:
    print(i.text + '--------' + i.label_ + '---------' + str(spacy.explain(i.label_)))
```

```python
# To print noun chunks of data
```

```python
for i in doc1.noun_chunks:
    print(i, ' ' * 8, i.text)
```

```python
# To import displacy from spacy
```

```python
from spacy import displacy
```

```python
# To render and display - diagram 1
```

```python
displacy.render(doc1, style = 'dep', jupyter = True, options = {'distance': 110})
```

```python
# To render and display - diagram 2
```

```python
displacy.render(doc, style = 'ent', jupyter = True)
```
