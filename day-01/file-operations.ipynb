```python
# String formatting
```

```python
name = "Bryan"
dic = {'address' : 123}

print(f"My name is {name}")

print("My name is " + name)

print("My name is {x}".format(x = name))

print(f"My address is {dic['address']}")

print(f"My name is {name} and address is {dic['address']}")

for i in name:
    print(i, end = ' | ')
```

```python
# Another way to define dictionary
```

```python
students = {('roll', 'name', 'dep'), (5010, 'Beniyal', 'IT'), (5011, 'Bryan', 'IT'), (5012, 'Daniel', 'IT')}
```

```python
num = 1
for student in students:
    print(f"Student {num}:\nRoll - {student[0]}\nName - {student[1]}\nDepartment - {student[2]}\n")
    num = num + 1
```
