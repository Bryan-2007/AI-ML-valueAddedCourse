```python
# Import Natural Language Tool Kit - NLTK
```

```python
import nltk

from nltk.stem.porter import *
```

```python
# Create object for porter stemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
# Defining an array of words
```

```python
words = ['hi', 'hello', 'bye', 'Hallo', 'hey', 'Tschuss']
```

```python
# It gives the stem word (fundamental word of the given word / base word)
for i in words:
    print(i, '--->', p_stemmer.stem(i))
```
