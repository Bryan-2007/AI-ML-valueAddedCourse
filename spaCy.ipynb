```python
# Importing spaCy module
```

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
# Define nlp of a sentence
```

```python
doc = nlp(u'Elon Musk bought Twitter for $5 million and renamed it to X')
```

```python
# Access the tokens with info
```

```python
print("Text\t\tPOS\t\tDependency\tLemmatization")
for token in doc:
    print(token.text + '\t\t' + token.pos_ + '\t\t' + token.dep_ + '\t\t' + token.lemma_)
```

```python
print(nlp.pipeline)
```
