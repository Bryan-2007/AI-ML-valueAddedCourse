```python
# To import a library
```

```python
from PyPDF2 import PdfReader
```

```python
# To open a pdf file
```

```python
f = open('assets/Business_Proposal.pdf', 'rb')
```

```python
# To load pdf file into object
```

```python
pdf_reader = PdfReader(f)
print(len(pdf_reader.pages))
```

```python
# To print content of a page
```

```python
page_one = pdf_reader.pages[0]
content = page_one.extract_text()
print(content)
```

```python
# To check whether a text is present in a pdf
```

```python
sample = "holistic world"
print(sample in content)
```

```python
# Importing RE module (Regular Expression)
```

```python
import re
```

```python
# Searching using RE module 1
```

```python
pattern = 'world'
result = re.search(pattern, content)
print(result)
```

```python
# To see exact location
```

```python
result.span()
```

```python
# To print all occurences of a text
```

```python
result = re.findall('world', content)
print(result)
```

```python
# Searching 2
```

```python
result = re.search(r'-', content)
print(result)
```

```python
# To save a re format and use later
```

```python
re_format = re.compile(r'-')
```

```python
# Using the saved format
```

```python
result = re.search(re_format, content)
print(result)
```

```python
# Using group() function
```

```python
pattern = re.compile(r"^(.*)\n([\s\S]+)", re.MULTILINE)

m = pattern.search(content)

print("\nGroup 1:\n\n", m.group(1))
print("\nGroup 2:\n\n", m.group(2))

print("\nFull grouped:\n\n", m.group())
```

```python
# More from search() function
```

```python
text = "Business Proposal is something that comes from the heart of a business man. That is made of iron and vibranium. I'm speaking about the heart and not about the business man"
m = re.search(r"Proposal", text)
print(m)
```
